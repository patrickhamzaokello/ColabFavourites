```python
def engineer_fraud_features(df, user_baselines=None, risk_thresholds=None):
    """
    Create features for fraud detection and flag suspicious transactions.

    Parameters:
    df: DataFrame with columns [transaction_id, date, amount, user_id, type, is_fraud]
    user_baselines: dict with user-specific thresholds (optional)
    risk_thresholds: dict with risk scoring thresholds (optional)

    Returns:
    DataFrame with engineered features and fraud flags
    """
    # Convert date to datetime if not already
    df['date'] = pd.to_datetime(df['date'])

    # Create copy to avoid modifications to original data
    features = df.copy()

    # [Previous feature engineering code remains the same until the end]
    user_stats = df.groupby('user_id').agg({
        'amount': ['mean', 'std', 'max', 'count'],
        'type': lambda x: x.value_counts().index[0]  # most common type
    }).reset_index()

    user_stats.columns = ['user_id', 'user_avg_amount', 'user_std_amount',
                         'user_max_amount', 'user_transaction_count',
                         'user_preferred_type']

    features = features.merge(user_stats, on='user_id', how='left')

    # Time-based features
    features['hour'] = features['date'].dt.hour
    features['day_of_week'] = features['date'].dt.dayofweek
    features['is_weekend'] = features['day_of_week'].isin([5, 6]).astype(int)
    features['is_night'] = ((features['hour'] >= 23) | (features['hour'] <= 4)).astype(int)

    # Amount-based features
    features['amount_vs_user_avg'] = features['amount'] / features['user_avg_amount']
    features['amount_vs_user_max'] = features['amount'] / features['user_max_amount']
    features['is_amount_outlier'] = (
        features['amount'] > features['user_avg_amount'] + 2 * features['user_std_amount']
    ).astype(int)

    # Transaction pattern features
    features['time_since_last_tx'] = features.groupby('user_id')['date'].diff().dt.total_seconds() / 3600

    def count_recent_tx(group, hours=24):
        # Reset index to access 'date' as a column
        group = group.reset_index(drop=True)
        def count_for_row(row):
            time_window = row['date'] - timedelta(hours=hours)
            return len(group[(group['date'] <= row['date']) &
                           (group['date'] > time_window)])
        return group.apply(count_for_row, axis=1)


    # features['tx_last_24h'] = features.groupby('user_id').apply(
    #     lambda x: count_recent_tx(x)
    # ).reset_index(level=0, drop=True)

    features['tx_last_24h'] = features.groupby('user_id').apply(
        lambda x: count_recent_tx(x)
    ).values

    # Type-based features
    features['is_preferred_type'] = (
        features['type'] == features['user_preferred_type']
    ).astype(int)

    # Amount velocity
    def amount_velocity(group, hours=24):
        # Reset index to access 'date' as a column
        group = group.reset_index(drop=True)
        def velocity_for_row(row):
            time_window = row['date'] - timedelta(hours=hours)
            recent_amount = group[(group['date'] <= row['date']) &
                                (group['date'] > time_window)]['amount'].sum()
            return recent_amount / (group['user_avg_amount'].iloc[0] * hours/24)
        return group.apply(velocity_for_row, axis=1) # Apply to each row

    # features['amount_velocity_24h'] = features.groupby('user_id').apply(
    #     lambda x: amount_velocity(x)
    # ).reset_index(level=0, drop=True)

    features['amount_velocity_24h'] = features.groupby('user_id').apply(
        lambda x: amount_velocity(x)
    ).values

    # Use default risk thresholds if none provided
    if risk_thresholds is None:
        risk_thresholds = {
            'amount_multiplier': 3,      # transactions > 3x user average
            'velocity_threshold': 2,      # velocity > 2x daily average
            'tx_frequency_threshold': 5,  # > 5 transactions in 24h
            'min_risk_score': 0.7        # minimum risk score to flag as fraud
        }

    # Calculate risk scores (0 to 1) for different aspects
    features['amount_risk'] = 1 - (1 / (1 + np.exp(features['amount_vs_user_avg'] - risk_thresholds['amount_multiplier'])))
    features['velocity_risk'] = 1 - (1 / (1 + np.exp(features['amount_velocity_24h'] - risk_thresholds['velocity_threshold'])))
    features['frequency_risk'] = 1 - (1 / (1 + np.exp(features['tx_last_24h'] - risk_thresholds['tx_frequency_threshold'])))
    features['time_risk'] = features['is_night'] * 0.3 + features['is_weekend'] * 0.2
    features['type_risk'] = (1 - features['is_preferred_type']) * 0.5

    # Calculate overall risk score
    features['risk_score'] = (
        features['amount_risk'] * 0.35 +    # Amount anomalies weighted highest
        features['velocity_risk'] * 0.25 +  # Velocity is next most important
        features['frequency_risk'] * 0.20 + # Frequency is third
        features['time_risk'] * 0.10 +      # Time-based risks weighted lower
        features['type_risk'] * 0.10        # Type-based risks weighted lower
    )

    # Flag transactions as potentially fraudulent based on risk score
    features['is_suspicious'] = (features['risk_score'] >= risk_thresholds['min_risk_score']).astype(int)

    # Add user baseline violations if provided
    if user_baselines:
        for user_id, baseline in user_baselines.items():
            mask = features['user_id'] == user_id
            if 'amount_threshold' in baseline:
                features.loc[mask, 'exceeds_user_threshold'] = (
                    features.loc[mask, 'amount'] > baseline['amount_threshold']
                ).astype(int)
                # Automatically flag as suspicious if user threshold is exceeded
                features.loc[mask & (features['exceeds_user_threshold'] == 1), 'is_suspicious'] = 1

            if 'allowed_types' in baseline:
                features.loc[mask, 'is_allowed_type'] = (
                    features.loc[mask, 'type'].isin(baseline['allowed_types'])
                ).astype(int)
                # Automatically flag as suspicious if transaction type not allowed
                features.loc[mask & (features['is_allowed_type'] == 0), 'is_suspicious'] = 1

    # Drop intermediate columns
    features = features.drop(['date', 'user_preferred_type'], axis=1)

    return features
```

```python
# Custom risk thresholds
risk_thresholds = {
    'amount_multiplier': 2.5,      # More sensitive to amount
    'velocity_threshold': 1.5,     # More sensitive to velocity
    'tx_frequency_threshold': 4,   # Flag after 4 transactions in 24h
    'min_risk_score': 0.6         # Lower threshold for suspicious flag
}

# With user-specific rules
user_baselines = {
    'user_id': {
        'amount_threshold': 1000,
        'allowed_types': ['bank_transfer']
    }
}

features = engineer_fraud_features(
    selected_df,
    user_baselines=user_baselines,
    risk_thresholds=risk_thresholds
)
```
